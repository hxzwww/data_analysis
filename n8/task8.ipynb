### Задача 1.

Найдите оценку параметра $\theta$ методом моментов по выборке размера $n$ из распределения:
* $U(a, b), \theta=(a, b)$; 
* $Pois(\theta)$;
* $\mathcal{N}(a, \sigma^2), \theta=(a, \sigma)$.

Выводить характеристики вероятностных распределений (мат. ожидание, дисперсия) не требуется.

<br>
<br>

Общая формула для нахождения оценки $\theta$:
$
\begin{cases}
E_\theta X_1 = \overline{X} \\
E_\theta X_1^2 = \overline{X^2} \\
\dots \\
E_\theta X_1^d = \overline{X^d}
\end{cases}
$
Где $X_1 \sim P_\theta$, $\theta \in \mathbf{R}^d$

<br>
<br>

##### №1
(d=2)
Для равномерного распределения значем, что
$
\begin{cases}
E_\theta X_1 = \frac{a + b}{2} \\
E_\theta X_1^2 =\frac{a^2 + ab + b^2}{3}\\
\end{cases}
\\
\ \ \ \ \ \ \ \ \ \ \ \Downarrow
\\
\begin{cases}
\frac{a + b}{2} = \overline{X} \\
\frac{a^2 + ab + b^2}{3} = \overline{X^2} \\
\end{cases}
\\
\begin{cases}
a = 2\overline{X} - b \\
a^2 + ab + b^2 = 3\overline{X^2} \\
\end{cases}
\\
(2\overline{X} - b)^2 + (2\overline{X} - b)b + b^2 = 3\overline{X^2}
\\
4\overline{X}^2 - 4\overline{X}b + b^2 + 2\overline{X}b - b^2 + b^2 = 3\overline{X^2}
\\
4\overline{X}^2 - 2\overline{X}b + b^2 = 3\overline{X^2}
\\
D = 4\overline{X}^2 - 4(4\overline{X}^2  - 3\overline{X^2}) = 12(\overline{X^2} - \overline{X}^2)
\\
b = \overline{X} \pm \sqrt{3(\overline{X^2} - \overline{X}^2)}
$
т.к. система была симметрична относитьльно a и b,
$
\widehat{a} = \overline{X} - \sqrt{3(\overline{X^2} - \overline{X}^2)}
\\
\widehat{b} = \overline{X} + \sqrt{3(\overline{X^2} - \overline{X}^2)}
\\
\widehat{\theta} = \Bigr(\overline{X} - \sqrt{3(\overline{X^2} - \overline{X}^2)},\ \overline{X} + \sqrt{3(\overline{X^2} - \overline{X}^2)} \Bigl)
$

<br>

##### №2
(d=1)
Для пуассоновского распределения значем, что
$
E_\theta X_1 = \theta
\\
\ \ \ \ \ \Downarrow
\\
\widehat{\theta}= \overline{X}
$

<br>

##### №2
(d=2)
Для нормального распределения значем, что
$
\begin{cases}
E_\theta X_1 = a \\
D_\theta X_1 = \sigma^2 \\
\end{cases}
\\
\ \ \ \ \ \Downarrow
\\
\widehat{a} = \overline{X}
\\
E_\theta X_1^2 = D_\theta X_1 + (E_\theta X_1)^2
\\
\overline{X^2} = \sigma^2 + a^2
\\
\widehat{\sigma} = \sqrt{\overline{X^2} - \overline{X}^2}
\\
\widehat{\theta} = \Bigr(\overline{X},\ \sqrt{\overline{X^2} - \overline{X}^2} \Bigl)
$

### Задача 3.   

**a).** Пусть $X_1,...,X_n$ выборка из некоторого распределения $\mathsf{P}$, причем ${\sf D} X_1=\sigma^2<+\infty$, и $\sigma$ неизвестно. 
Рассмотрим оценку $S^2 = \overline{X^2}-\overline{X}^2$ дисперсии $\sigma^2$.

Докажите, что $S^2=\frac1n\sum\limits_{i=1}^n\left(X_i-\overline{X}\right)^2$.

<br>
<br>
<br>

$
\overline{X} = \sum_{i=1}^n \frac{1}{n} X_i
\\
S^2  =  \overline{X^2} - \overline{X}^2
 =
 \overline{X^2} - 2\overline{X}^2 + \overline{X}^2
 =
 \sum_{i=1}^n \frac{1}{n} X_i^2 - 2(\sum_{i=1}^n \frac{1}{n} X_i)^2 + (\sum_{i=1}^n \frac{1}{n} X_i)^2
 =
 \frac{1}{n^3} n^2 \sum_{i=1}^n X_i^2 - \frac{2}{n^3} n \sum_{i=1}^n (X_i \sum_{j=1}^n X_j) + \frac{1}{n^3} n (\sum_{i=1}^n X_i)^2
 =
\\ =  \frac{1}{n^3} \Bigr( \sum_{i=1}^n (n X_i)^2 - 2\sum_{i=1}^n (n X_i \sum_{j=1}^n X_j) + \sum_{i=1}^n (\sum_{j=1}^n X_j)^2 \Bigl)  =
 \frac{1}{n^3} \sum_{i=1}^n \Bigr(n X_i - \sum_{j=1}^n X_j \Bigl)^2  =
 \frac{1}{n} \sum_{i=1}^n \Bigr(X_i - \sum_{j=1}^n \frac{1}{n} X_j \Bigl)^2  =
 \frac{1}{n} \sum_{i=1}^n \Bigr(X_i - \overline{X} \Bigl)^2
$

<br>
<br>
<br>

**b).** Оценка $\widehat{\theta}$ называется несмещенной оценкой параметра $\theta$ если для любого $\theta \in \Theta$ выполнено $\mathsf{E}_\theta \widehat{\theta} = \theta$. Иначе говоря, какое бы ни оказалось истинное значение параметра $\theta$ рассматривая оценку $\widehat{\theta}$ в среднем будем получать именно $\theta$.

Является ли статистика $S^2$ несмещенной оценкой $\sigma^2$?

*Подсказка.* Посчитайте математическое ожидание случайной величины $S^2$. Используйте для этого известные вам свойства математического ожидания.

<br>
<br>
<br>

Посчитаем $ES^2$
$
ES^2 = E(\overline{X^2} - \overline{X}^2) = E\overline{X^2} - E\overline{X}^2 = EX_1^2 - E\overline{X}^2
\\
\sigma^2 = DX_1 = EX_1^2 - (EX_1)^2 \Rightarrow ES^2 = \sigma^2 \iff E\overline{X}^2 = (EX_1)^2
$


$E \overline{X}^2 = \frac{1}{n^2} \sum_{i=1, j=1}^{n, n} E(X_i * X_j) =
\frac{1}{n^2} \Bigl(\sum_{i=1, j=1, i \neq j}^{n, n} E(X_i * X_j) +
\sum_{i=1}^{n} E(X_i^2)\Bigr) =
\frac{1}{n^2} \Bigl(\sum_{i=1, j=1, i \neq j}^{n, n} EX_i * EX_j) +
\sum_{i=1}^{n} E(X_i^2)\Bigr) = \\
 = \frac{1}{n^2} \Bigl(\sum_{i=1, j=1, i \neq j}^{n, n} (EX_1)^2 +
\sum_{i=1}^{n} E(X_1^2)\Bigr) = \frac{1}{n^2} \Bigl(n(n-1)(EX_1)^2+nEX_1^2\Bigr) = (EX_1)^2 + \frac{1}{n}\Bigl(EX_1^2-(EX_1)^2\Bigr) = (EX_1)^2 + \frac{\sigma^2}{n}$

$E\overline{X}^2 \ne (EX_1)^2 \Rightarrow S^2$ не является несмещенной оценкой $\sigma^2$
$
ES^2 = EX_1^2 - (EX_1)^2 - \frac{\sigma^2}{n} = \sigma^2 - \frac{\sigma^2}{n} = \frac{n-1}{n}\sigma^2
$

### Задача 4.   
Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \sigma^2)$. В качестве оценок $\sigma^2$ обычно используют $S^2$ или $\frac{n}{n-1}S^2$. Что можно сказать про их несмещенность?

**Ответ:**
$DX_1 = \sigma^2$
Из 3 задачи знаем что $S^2$ не является несмещенной оценкой
$E(\frac{n}{n-1}S^2) = \frac{n}{n-1}ES^2 = \sigma^2 \Rightarrow \frac{n}{n-1}S^2$ является несмещенной оценкой

**Решение:**

#### $\sigma := 1$

```python
def generate(_n, _k, _sigma=1, _estimation_type=0):
    xin = stats.norm.rvs(loc=0, scale=_sigma, size=(_k, _n))
    thetas = (
                     np.mean(np.power(xin, 2), axis=1) - np.power(np.mean(xin, axis=1), 2)
             ) * (
                 _n / (_n - 1) if _estimation_type else 1
             )
    return thetas, np.mean(thetas)


k = 500
n_range = [10, 100, 500]
est_type_range = [0, 1]
```

```python
data = []

for n in n_range:
    data_n = []
    for est_type in est_type_range:
        data_n.append(
            (generate(n, k, 1, est_type))
        )
    data.append(data_n)
```

```python
colors = ['yellow', 'green']
labels = ['$S^2$', r'$\frac{n}{n-1} S^2$']

plt.figure(figsize=(15, 5))
with sns.plotting_context(font_scale=1.5), sns.axes_style("darkgrid"):
    for i, data_n in enumerate(data):
        plt.subplot(1, 3, i + 1)
        for y, color, label in zip(est_type_range, colors, labels):
            plt.scatter(data_n[y][0], np.zeros(k) + y,
                        alpha=0.1, s=100, color=color, label=label)
            plt.scatter(data_n[y][1], y, marker='*', s=200,
                        color='w', edgecolors='black')

        plt.vlines(1, 0, 1, color='r')
        plt.title(f'Размер выборки = {n_range[i]}')
        plt.yticks([])
        plt.legend()
```

**2.** Изучим поведение среднего оценок из первого пункта при росте размера выборки. Постройте график зависимости $\widetilde{\theta}$ от $n$ для двух типов оценок. Для вычисления зависимости нужно один раз сгенерировать выборки из пункта 1 достаточно большого размера (не более 500) и посчитать оценки по подвыборкам, используя <a href="https://miptstats.github.io/courses/python/05_numpy.html#2.1-Математические-операции">функции</a> из `numpy`. Использовать циклы, а так же функции, разворачивающиеся в цикл (например, `np.vectorize`), запрещено. 

**Решение:**

```python
n = 300
k = 300
X = stats.norm.rvs(loc=0, scale=1, size=(k, n))
data = []

n_range = np.arange(1, n + 1)
nm1_range = n_range - 1
nm1_range[0] = 1

for est_type in est_type_range:
    theta = (
                    np.mean(np.power(X, 2), axis=1) - np.power(np.mean(X, axis=1), 2)
            ) * (
                n_range / nm1_range if est_type else np.ones(k)
            )
    data.append(theta.cumsum() / n_range)
```

```python
plt.figure(figsize=(9, 9))
for i, averages in enumerate(data):
    plt.plot(n_range, averages, label = labels[i], color=colors[i])
plt.legend()
plt.title("Зависимость оценки от размера выборки")
plt.ylabel(r"$\widetilde{\theta}$ - средняя оценка")
plt.xlabel("Размер выборки")
plt.show()
```
